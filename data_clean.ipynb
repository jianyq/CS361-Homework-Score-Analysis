```python
# data_total.describe()
for i in range(1, 11):
    df = pd.read_csv(f'CS361_CourseDev_Trend\HW{i}_clean.csv')
    df = df[["randNumCol", "Total Score"]]
    # only keep one row for each randNumCol
    df = df.groupby("randNumCol").mean().reset_index()
    # rename the columns of the dataframe
    df.columns = ["randNumCol", f"HW{i}"]

    # merge this dataframe with data_total on "randNumCol"
    # directly let HW1 = df["HW1"] where randNumCol is the same
    data_total = pd.merge(data_total, df, on="randNumCol", how="left")
# save to csv file
data_total.to_csv(f'CS361_CourseDev_Trend\data_total.csv')
```
